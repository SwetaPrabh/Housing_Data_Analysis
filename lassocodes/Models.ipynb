# Penalized Logistic Regression(Lasso)


```python
#Reading the test and train Datasets
train_final = pd.read_csv('train_final.csv', index_col=0)
test_final = pd.read_csv('test_final.csv', index_col =0)
```

```python
train_final=train_final.reset_index()
test_final=test_final.reset_index()
```

```python
#train_x, train_y
y_train = train_final['SalePrice']
X_train = train_final.drop(['SalePrice','PID'], axis=1)
```

```python
#test_x, test_y
y_test = test_final['SalePrice']
X_test = test_final.drop(['SalePrice','PID'], axis=1)
```

```python
X_train[:] = np.nan_to_num(X_train)
X_test[:] = np.nan_to_num(X_test)
```

```python
lasso  = Lasso()
alphas = np.linspace(0.7,40.7,41)
lasso.set_params(normalize=True)
coefs_lasso  = []

for alpha in alphas:
        lasso.set_params(alpha=alpha, normalize=True)
        lasso.fit(X_train, y_train)  
        
        coefs_lasso.append(lasso.coef_)
        print('Accuracy of lasso with lambda {:.2f} on train set: {:.2f}'.format(alpha, lasso.score(X_train, y_train)))
        print('Accuracy of lasso on test set: {:.2f}'.format(lasso.score(X_test, y_test)))



coefs_lasso = pd.DataFrame(coefs_lasso, index = alphas, columns = X_train.columns)  
coefs_lasso = coefs_lasso.reset_index()
coefs_lasso
```

```python
# with alpha 0.7, the train accuracy is 0.96 and test accuracy is 0.92
lasso  = Lasso()
alpha = 0.7
lasso.set_params(normalize=True)
coefs_lasso_final  = []


lasso.set_params(alpha=0.7, normalize=True)
lasso.fit(X_train, y_train)  
        
coefs_lasso_final.append(lasso.coef_)
print('Accuracy of lasso with lambda {:.2f} on train set: {:.2f}'.format(alpha, lasso.score(X_train, y_train)))
print('Accuracy of lasso on test set: {:.2f}'.format(lasso.score(X_test, y_test)))



coefs_lasso1 = pd.DataFrame(coefs_lasso_final, columns = X_train.columns)  
coefs_lasso1 = coefs_lasso1.reset_index()
# coefs_lasso
```

```python
from sklearn.svm import SVR
svr = SVR(epsilon = 1e-4, gamma=1)
svr.fit(X_train,y_train)
print(svr.score(X_train,y_train))
print(svr.score(X_test,y_test))
```

```python
R2 = []
n_support_epsilon = []
support_vectors_epsilon = []
epsilons = np.logspace(1,-1,20)
for epsilon in epsilons:
    svr.set_params(epsilon=epsilon, gamma=1)
    svr.fit(X_train,y_train)
    R2.append(svr.score(X_test,y_test))
    n_support_epsilon.append(svr.support_vectors_.shape[0])
    support_vectors_epsilon.append(svr.support_vectors_)

n_support_epsilon = np.array(n_support_epsilon)
support_vectors_epsilon = np.array(support_vectors_epsilon)
R2    
```

```python
#Gridsearch on gamma and epsilon
```

```python
M, M2 = 12, 12

gammas = np.logspace(3,-3,M)
epsilons = np.logspace(3,-3,M2)
scores = []
train_score = []
test_score = []
gammalist = []
epsilonlist =[]
for gamma in gammas:
    for epsilon in epsilons:
            svr.set_params(gamma=gamma, epsilon=epsilon)
            svr.fit(X_train,y_train)
            gammalist.append(gamma)
            epsilonlist.append(epsilon)
            scores.append(svr.score(X_train,y_train)-svr.score(X_test,y_test)) # train R^2 and test R^2 gaps
            train_score.append(svr.score(X_train,y_train))
            test_score.append(svr.score(X_test,y_test))
            
scores1 = np.array(scores).reshape((M, M2))            
```
