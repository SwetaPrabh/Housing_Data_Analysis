# Random Forest 

```python
# Pandas is used for data manipulation
import pandas as pd

# Read in train data as pandas dataframe 
features_train = pd.read_csv('train_final.csv')
tmp_groupby_df = features_train.groupby(['YearBuilt']).agg(SalePrice_mean=('SalePrice', 'mean')).reset_index()
features_train = features_train.merge(tmp_groupby_df, on=['YearBuilt'])
features_train
```

```python
print('Shape of features after one-hot encoding:', features_train.shape)
```

```python
# Read in test data as pandas dataframe and display first 5 rows
features_test = pd.read_csv('test_final.csv')
tmp_groupby_df = features_test.groupby(['YearBuilt']).agg(SalePrice_mean=('SalePrice', 'mean')).reset_index()
features_test = features_test.merge(tmp_groupby_df, on=['YearBuilt'])
features_test["BsmtFullBath"].fillna(0, inplace = True)
features_test["BsmtHalfBath"].fillna(0, inplace = True)
features_test
```

```python
import bamboolib
```

```python
# Use numpy to convert to arrays
import numpy as np

# Labels are the values we want to predict
labels_train = np.array(features_train['SalePrice'])
labels_test = np.array(features_test['SalePrice'])

# Remove the labels from the features
# axis 1 refers to the columns
features_train = features_train.drop('SalePrice', axis = 1)
features_test = features_test.drop('SalePrice', axis = 1)

# Saving feature names for later use
feature_train_list = list(features_train.columns)
feature_test_list = list(features_test.columns)

# Convert to numpy array
features_train = np.array(features_train)
features_test = np.array(features_test)
```

```python
# The baseline predictions are the SalePrice averages by YrSold
baseline_preds = features_test[:, feature_test_list.index('SalePrice_mean')]

# Baseline errors, and display average baseline error
baseline_errors = abs(baseline_preds - labels_test)
print('Average baseline error: ', round(np.mean(baseline_errors), 2), '$.')
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor

# Instantiate model 
rf = RandomForestRegressor(n_estimators= 1000, random_state=42)

# Train the model on training data
rf.fit(features_train, labels_train);
```

```python
rf.score(features_train, labels_train)
```

```python
#Overfitting
rf.score(features_test, labels_test)
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(features_test)

# Calculate the absolute errors
errors = abs(predictions - labels_test)

# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), '$.')
```

# Let the Hyperparameter Tuning Fun Begin! 

# Random Hyperparameter Grid

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
# Random search of parameters, using 5 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 5, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(features_train, labels_train)
```

```python
def evaluate(model, features_test, labels_test):
    predictions = model.predict(features_test)
    errors = abs(predictions - labels_test)
    mape = 100 * np.mean(errors / labels_test)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} $.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    
    return accuracy

base_model = RandomForestRegressor(n_estimators = 10, random_state = 42)
base_model.fit(features_train, labels_train)
base_accuracy = evaluate(base_model, features_test, labels_test)
```

```python
best_random = rf_random.best_estimator_
random_accuracy = evaluate(best_random, features_test, labels_test)
```

```python
print('Improvement of {:0.2f}%.'.format( 100 * (random_accuracy - base_accuracy) / base_accuracy))
```

```python
from sklearn.model_selection import GridSearchCV
# Create the parameter grid based on the results of random search 
param_grid = {
    'bootstrap': [False],
    'max_depth': [120, 130, 140, None],
    'max_features': ['sqrt', 100, 200, 230, 240],
    'min_samples_leaf': [1, 2, 3],
    'min_samples_split': [1, 2, 3],
    'n_estimators': [300, 350, 400, 450]
}
# Create a based model
rf = RandomForestRegressor()
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 5, n_jobs = -1, verbose = 2)
```

```python
grid_search.fit(features_train, labels_train)
grid_search.best_params_
```

```python
best_grid = grid_search.best_estimator_
grid_accuracy = evaluate(best_grid, features_test, labels_test)
print('Improvement of {:0.2f}%.'.format( 100 * (grid_accuracy - base_accuracy) / base_accuracy))
```
